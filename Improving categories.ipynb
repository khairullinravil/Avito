Построим удобную иерархическую структуру для DataFrame categories:

Добавим еще несколько столбцов для хранения каждого подкласса:

```python
max_subcategories = 0
for name in categories['name']:
    sep_count = 0
    for i in name:
        if (i == '|'):
            sep_count += 1
    max_subcategories = max(max_subcategories, sep_count)
for i in range(1, max_subcategories + 1):
    categories[(i * 'sub') + 'name'] = 0
```

```python
categories.head()
```

```python
for i, name in enumerate(categories['name']):
    subnames = name.split(sep='|')
    for j, substring in enumerate(subnames):
        categories[j * 'sub' + 'name'][i] = substring
```
